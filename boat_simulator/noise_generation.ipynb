# Noise Function

```python
import numpy as np
import matplotlib.pyplot as plt

T = 100
x = np.arange(0, T)
clean_signal = np.sin(x)
noise = np.random.normal(0, 1, T)

fig, axs = plt.subplots(5)

axs[1].plot(noise)
axs[1].set_ylabel("noise value")

axs[0].plot(clean_signal)
axs[0].set_ylabel("signal value")

noisy_signal = noise + clean_signal
axs[2].plot(noise + clean_signal)
axs[2].set_ylabel("noisy value")

epsilon = 0.05
peak_value = 10
coin_flips = np.random.choice([1, 0], T, p=[epsilon, 1 - epsilon])

axs[3].plot(coin_flips)
axs[3].set_ylabel("coin flips")

noisy_spiked_signal = noisy_signal + (coin_flips * peak_value)
axs[4].plot(noisy_spiked_signal)
axs[4].set_ylabel("signal value")

plt.show()
```
